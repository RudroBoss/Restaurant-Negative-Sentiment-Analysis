```python
import pandas as pd

# Load dataset (replace 'your-dataset-path' with actual path from Kaggle dataset)
df = pd.read_csv('/kaggle/input/restaurant-review-dataset/Restaurant reviews.csv')

# Display first 5 rows
df.head()
```

```python
df.columns
```

```python

df = df.drop(columns=['7514'])

df.head()
```

```python
# Show data types of each column
df.dtypes
```

```python
from textblob import TextBlob

# Function to get sentiment polarity safely
def get_sentiment(text):
    if isinstance(text, str):  # Check if text is a string
        return TextBlob(text).sentiment.polarity
    else:
        return 0  # Neutral sentiment for missing/invalid values

# Apply sentiment analysis
df['Sentiment'] = df['Review'].apply(get_sentiment)

# Keep only negative reviews (polarity < 0)
df_negative = df[df['Sentiment'] < 0]

# Show negative reviews
df_negative.head()
```

```python
# Count the number of negative reviews
num_negative_reviews = df_negative.shape[0]

# Display the result
print("Total negative reviews:", num_negative_reviews)
```

```python
# Convert 'Rating' to numeric (if not already)
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

# Keep only negative sentiment reviews with a rating of 1 or 2
df_negative = df[(df['Sentiment'] < 0) & (df['Rating'].isin([1, 2]))]

# Show the first few rows
df_negative.head()
```

```python
# Count the number of negative reviews
num_negative_reviews = df_negative.shape[0]

# Display the result
print("Total negative reviews:", num_negative_reviews)
```

```python
# Save negative reviews to a CSV file
df_negative.to_csv('negative_reviews.csv', index=False)

print("File saved as negative_reviews.csv")
```
